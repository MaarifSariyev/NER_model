```python
import random
import json

names = ["Aytac Məmmədova", "Rəşad Əliyev", "Nərmin İsmayılova", "Elvin Qasımov","Elmir Əliyev","Kənan Vəliyev","Vaqif Qedimov","Murad Muradov"]
cities = ["Bakı", "Sumqayıt", "Gəncə", "Şəki", "Quba", "Xırdalan","Qazax","Kürdəmir","Ucar"]
organizations = ["Kapital Bank ASC", "Dövlət Gömrük Komitəsi", "TexnoPlus MMC","Milli Bank","Az Telecom","Holberton School"]
emails = ["aytac.mammadova@mail.az", "rashad.aliyev@example.com","nermin.ismaylova1@gmail.com","qasimov33@gmail.com","elmir.eliyev5@gmail.com","veliyev.kenan@gmail.com","vaqif.qdmov0@gmail.com","murad.muradov@gmail.com"]
phones = ["+994 50 123 45 67", "+994 99 698 74 44", "+994 50 261 69 85", "+994 99 428 34 14", "+994 55 908 88 90", "+994 70 691 78 50", "+994 77 536 25 77", "+994 55 574 08 90"]
ids = ["AZE76439821", "AZE99348274", "AZE56783219", "AZE13547486", "AZE28446580", "AZE67894328"]
tax_ids = ["1801234567", "2007654321"]
accounts = ["AZ21NABZ01350100000000001944", "AZ33ACDB00000000000000001234"]
card_suffixes = ["1234", "9876", "3456"]
dates = ["14 mart 1985-ci il", "25 aprel 1991-ci il", "10 yanvar 1990-cı il"]
nationalities = ["Azərbaycan Respublikası", "Türkiyə Respublikası"]

def tokenize(text):
    return text.split()

def tag_tokens(tokens, tag_type):
    tags = []
    for i, _ in enumerate(tokens):
        tags.append(f"B-{tag_type}" if i == 0 else f"I-{tag_type}")
    return tags

def create_sample():
    name = random.choice(names)
    city = random.choice(cities)
    org = random.choice(organizations)
    email = random.choice(emails)
    phone = random.choice(phones)
    id_number = random.choice(ids)
    tax_id = random.choice(tax_ids)
    account = random.choice(accounts)
    card = "**** **** **** " + random.choice(card_suffixes)
    date = random.choice(dates)
    nationality = random.choice(nationalities)

    tokens = []
    ner_tags = []

    # Add name
    tokens += tokenize(name)
    ner_tags += tag_tokens(tokenize(name), "PERSON")

    # Add date
    tokens += tokenize(date)
    ner_tags += tag_tokens(tokenize(date), "DATE")

    # ID
    tokens += tokenize(id_number)
    ner_tags += tag_tokens(tokenize(id_number), "ID")

    # Nationality
    tokens += tokenize(nationality)
    ner_tags += tag_tokens(tokenize(nationality), "NATIONALITY")

    # Phone
    tokens += tokenize(phone)
    ner_tags += tag_tokens(tokenize(phone), "PHONE")

    # Email
    tokens += tokenize(email)
    ner_tags += tag_tokens(tokenize(email), "EMAIL")

    # Location
    tokens += tokenize(city)
    ner_tags += tag_tokens(tokenize(city), "LOCATION")

    # Organization
    tokens += tokenize(org)
    ner_tags += tag_tokens(tokenize(org), "ORGANIZATION")

    # Tax ID
    tokens += tokenize(tax_id)
    ner_tags += tag_tokens(tokenize(tax_id), "TAX_ID")

    # Account
    tokens += tokenize(account)
    ner_tags += tag_tokens(tokenize(account), "ACCOUNT")

    # Card
    tokens += tokenize(card)
    ner_tags += tag_tokens(tokenize(card), "CARD")

    return {"tokens": tokens, "ner_tags": ner_tags}

# Generate 1000 samples
data = [create_sample() for _ in range(2000)]

# Save to file
with open("kyc_ner_dataset.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
